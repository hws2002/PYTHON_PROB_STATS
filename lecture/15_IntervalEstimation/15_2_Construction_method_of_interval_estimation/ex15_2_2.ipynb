```python
from scipy.stats import chi2

n = 10000 # Sample size (= degree of freedom+1 for chi-square distribution)

S = 20 # Sample standard deviation

# confidence level
confidence_level = 0.95
alpha = 1 - confidence_level
```

```python
# calculate critical value of the chi-square distribution
# where confidence level is `confidence_level` and degree of freedom is `n-1`

critical_value_r = chi2.ppf(1 - alpha/2, n-1)

print(f"The critical value for chi^2_(1-a/2)(n) with a = {alpha} and n = {n} is: {critical_value_r}")
```

```python
critical_value_l = chi2.ppf(alpha/2, n-1)

print(f"The critical value for chi^2_(a/2)(n) with a = {alpha} and n = {n} is: {critical_value_l}")
```

```python
print(f"Required Confidence interval is [((n-1)S^2)/X^2_(1-a/2)(n-1) ,((n-1)S^2)/X^2_(a/2)(n-1) ] =[{(n-1)*(S**2)/critical_value_r},{(n-1)*(S**2)/critical_value_l}]")
```
