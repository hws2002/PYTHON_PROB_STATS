## 卡方分布的概率分布函数

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Define the degrees of freedom
df = 15

# Generate x-axis values from 0 to 40
x = np.linspace(0, 40, 500)

# Calculate the corresponding probability density function (PDF) values
cdf_15 = chi2.cdf(x, df)

# Plot the chi-square distribution
plt.plot(x, cdf_15, label=f"Chi-square({df})")

# Set the plot title and labels
plt.title(f"Chi-square Distribution (df={df})")
plt.xlabel("x")
plt.ylabel("Cumulative Density Function (CDF)")

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

```python
# Generate x-axis values from 0 to 20
x = np.linspace(0, 40, 500)

df_list = [4,6,10,20]
for df in df_list:
    # Calculate the corresponding probability density function (PDF) values
    cdf = chi2.cdf(x, df)
    # Plot the chi-square distribution
    plt.plot(x, cdf, label=f"Chi-square({df})")

# Set the plot title and labels

# Convert the values in df_list to strings
df_values = [str(df) for df in df_list]

# Join the values inside curly brackets
plt.title("Chi-square Distribution (df = {})".format(", ".join(df_values)))
plt.xlabel("x")
plt.ylabel("Cumulative Density Function (PDF)")

# Show the legend
plt.legend()

# Display the plot
plt.show()
```
